### Collect Data using tweepy Cursor

**We are going to collect data in europe & USA**

```python
geo_usa = "40.68908,-100.95860,1500km"
geo_europe = "51.18348,10.03954,1200km"
```

```python
def Collector(API_search, term, lang, number_of_tweets, geo, file_name):
    """
    :param API_search: 
    :param term: 
    :param lang: 
    :param number_of_tweets: 
    :param file_name: 
    :return: 
    """
    tweets = Cursor(API_search, count=100000, geocode=geo, q=term, lang=lang).items(number_of_tweets)
    print("collecting")
    with open(file_name + '.txt', 'a') as file:
        for elm in tweets:
            tweet = elm._json
            try:
                file.write(json.dumps(tweet) + '\n')
            except Exception as e:
                print('Unable to collect this tweet because of {}'.format(e))
                pass
            pass
```
